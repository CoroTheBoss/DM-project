```python
import pandas as pd
```

```python
df = pd.read_csv('integrated.csv', index_col=0)
df
```

```python
quality_esse = df[df['Esselunga category'] == 'latticini, salumi e formaggi'].dropna(subset=['Conad', 'Carrefour'],thresh=1).sample(50)
quality_esse.reset_index(drop=True, inplace=True)
quality_esse
```

```python
for i in range(len(quality_esse)):
    print(quality_esse.iloc[i]['Esselunga'])
    print(quality_esse.iloc[i]['Carrefour'])
    print(quality_esse.iloc[i]['Conad'])
    print('-------')
    
# 41/50
```
