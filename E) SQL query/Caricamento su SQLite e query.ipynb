**Query1**

```python
import csv


conn = sqlite3.connect('/Users/Chicca/Desktop/Integrated_final.db')
cursor = conn.cursor()
query = "SELECT * FROM Integrated WHERE LOWER(Carrefour) LIKE '%bonduelle%' OR LOWER(Conad) LIKE '%bonduelle%' OR LOWER(Esselunga) LIKE '%bonduelle%'"
cursor.execute(query)


results = cursor.fetchall()


csv_file = '/Users/Chicca/Desktop/Query1.csv'


with open(csv_file, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow([i[0] for i in cursor.description]) 
    writer.writerows(results) 

conn.close()
```

**Query2**

```python
conn = sqlite3.connect('/Users/Chicca/Desktop/Integrated_final.db')
cursor = conn.cursor()

query = """
SELECT *,
       CASE
           WHEN "Esselunga price" <= "Carrefour price" AND "Esselunga price" <= "Conad price" THEN "Esselunga price"
           WHEN "Carrefour price" <= "Esselunga price" AND "Carrefour price" <= "Conad price" THEN "Carrefour price"
           ELSE "Conad price"
       END AS "Most_Convenient"
FROM Integrated
WHERE (LOWER(Carrefour) LIKE '%noberasco%' OR LOWER(Conad) LIKE '%noberasco%' OR LOWER(Esselunga) LIKE '%noberasco%')
  AND (Carrefour IS NOT NULL AND Conad IS NOT NULL AND Esselunga IS NOT NULL)
  AND (LOWER("Carrefour category") LIKE '%frutta%' OR LOWER("Conad category") LIKE '%frutta%' OR LOWER("Esselunga category") LIKE '%frutta%')
"""

cursor.execute(query)


results = cursor.fetchall()


csv_file = '/Users/Chicca/Desktop/Query2.csv'


with open(csv_file, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow([i[0] for i in cursor.description]) 
    writer.writerows(results) 

conn.close()
```

**Query 3**

```python
conn = sqlite3.connect('/Users/Chicca/Desktop/Integrated_final.db')
cursor = conn.cursor()

query = """
SELECT MIN("Carrefour price") AS min_carrefour, MIN("Conad price") AS min_conad, MIN("Esselunga price") AS min_esselunga,
       MIN("Carrefour price") + MIN("Conad price") + MIN("Esselunga price") AS total
FROM Integrated
WHERE (LOWER("Carrefour") LIKE '%mela%' OR LOWER("Conad") LIKE '%mela%' OR LOWER("Esselunga") LIKE '%mela%')
   AND (LOWER("Carrefour") LIKE '%pera%' OR LOWER("Conad") LIKE '%pera%' OR LOWER("Esselunga") LIKE '%pera%')
   AND (LOWER("Carrefour") LIKE '%banana%' OR LOWER("Conad") LIKE '%banana%' OR LOWER("Esselunga") LIKE '%banana%');
 
"""

cursor.execute(query)


results = cursor.fetchall()


csv_file = '/Users/Chicca/Desktop/Query3.csv'


with open(csv_file, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow([i[0] for i in cursor.description]) 
    writer.writerows(results) 

conn.close()
```
